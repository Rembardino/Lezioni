## Spazio vettoriale

<font size="4"> Questo concetto  la base dell'algebra lineare. La definizione generale ci dice che uno spazio vettoriale  composto da un _campo_ , da un insieme di _vettori_ e da due operazioni: _somma_ e _prodotto per uno scalare_. <br> <br>
Le parole _vettori, somma e prodotto per uno scalare_ dovrebbero essere familiari se avete visto la lezione dove ho introdotto i vettori, se non lo sono vi consiglio di vedere quella parte della lezione per rinfrescacrvi un pochino la memoria. La parola nuova  _campo_.<br> <br>
In matematica il _campo_  un insieme di numeri e da due operazioni:somma e prodotto. Esempi di campi sono il campo dei numeri reali (campo R), il campo dei numeri razionali (Q) e il campo complesso (C).<br> <br>
Condizioni affinche uno spazio sia vettoriale:
* deve esistere un elemento nullo (lo 0)<br> <br>
* la somma  commutativa e associativa (quanto tempo che non ascoltavate queste parole vero?!)<br> <br>
* ogni vettore ha un opposto <br> <br>
    
Ve lo avevo detto che in questa lezioni ci sono molte nozioni teoriche. Per cercare di capire cosa sta succecdendo qui, proviamo a fare questo esempio: 
Siamo nella citt della matematica il sabato sera nella zona dei locali. Per entrare nella discoteca, bisogna che i numeri entrino insieme. Una volta creati questi gruppi i numeri possono entrare ma solamente al primo piano chiamato "campo". Per andare al piano succcecssivo, per esempio, devono unirsi alle propriet precedenti e quindi poter entrare e diventare uno spazio vettoriale. Ora da questo piano possono andare su molti altri piano del locale ma ogniuno di questi ha le sue regole da dover soddisfare. Si, la citt della matematica  molto selettiva

<font size="4"> Possiamo ora ritornare alla nostra definizione di applicazione lineare. <br> <br>
_Dati due spazi vettoriali V e W definiti su un campo K, e F una funzione da V a W, allora F si dice applicaczione lineare se le seguenti condizioni sono verificate:_ <br> <br>
* $F(v_1+v_2)=F(v_1)+F(v_2)$<br> <br>
* $F(\lambda v)= \lambda F(v)$<br> <br>
* il valore nullo 0 manda a zero la funzione $F(0)=0$<br> <br>
Ogni volta che leggerete lineare dovete sempre pensare a queste  condizioni!<br> <br>
<font size="5"> <b>Esempi:</b></font><br> <br>
* F definita $R \to R$   <br> <br>
    $F(x)=8(x)$  lineare perch $F(x+y)=8(x+y)\to 8x+8y = F(x)+F(y)$ e $F(0)=8*0=0$  <br> <br>
* F definita $R \to R$   <br> <br>
    $F(x)=8(x)+1$ non  lineare perch $F(0)=8*0+1=1$ 
<font size="5"> <b>Altri Esempi di applicazioni:</b></font><br> <br>    
* F definita $R^2 \to R$<br> <br>
    $F(x,y)=x-y^2$

# Le matrici

<font size="4"> Forse l'applicazione lineare piu usata  la matrice. In fisica quantistica le matrici vengono utilizzate per rappresentare le varie grandezze fisiche da studiare. Gli autovalori di queste matrici sono i risulati possibili della misura. Quindi, quando e se mai calcolerete gli autovalrori di una matrice, sappiate che praticamente state risolvendo un calcolo quantistico.  <br> <br>
Sulle propriet delle matrici si potrebbe parlare per settimane ma cerchiamo di nominare le pi importanti. 
* Prodotto di una matrice per un vettore:<br> <br>
    $\begin{pmatrix} a & b\\ c & d \end{pmatrix}$*$\begin{pmatrix} e\\ f \end{pmatrix}=\begin{pmatrix} (a*e)+(b*f)\\ (c*e)+(d*f) \end{pmatrix}$<br> <br>

<font size="5"> <b>Esempio:</b></font> <br><br>
    $\begin{pmatrix} 1 & 2\\ 3 & 4 \end{pmatrix}$*$\begin{pmatrix} 5\\ 6 \end{pmatrix}=\begin{pmatrix} (1*5)+(2*6)\\ (3*5)+(4*6) \end{pmatrix}=\begin{pmatrix} 18\\ 39 \end{pmatrix}$
    
* Prodotto di una matrice per una matrice:<br> <br>
 $\begin{pmatrix} a & b\\ c & d \end{pmatrix}$*$\begin{pmatrix} e & f\\ g & h \end{pmatrix}$=$\begin{pmatrix}(a*e+b*g) & (a*f+b*h)\\ (c*e+f*g) & (c*f+d*h) \end{pmatrix}$<br><br>
    
<font size="5"> <b>Esempio:</b></font> <br><br>
     $\begin{pmatrix} 1 & 2\\ 3 & 4 \end{pmatrix}$* $\begin{pmatrix} 1 & 0\\ 0 & 2 \end{pmatrix}$= $\begin{pmatrix} (1*1+2*0) & (1*0+2*2)\\ (3*1+4*0) & 3*0+4*2) \end{pmatrix}$= $\begin{pmatrix} 1 & 4\\ 3 & 8 \end{pmatrix}$

<font size="4"> Quando si ha a che fare con le matrici, purtroppo, le dimensioni contano. E anche parecchio!. Infatti non possiamo moltiplicare matrici se il numero delle colonne della prima  diverso dal numero di righe della seconda. <br><br>

Inoltre, l'ordine di moltiplicazione  importante! se ho due matrici A e B, A * B potrebbe avere un risultato di verso rispoetto a B * A! Provatelo con il secondo esempio invertendo l'ordine delle matrici


<font size="4"> Come mai la matrice  un'applicazione lineare? Bene, dato un vettore x, la matrice trasforma x in un nuovo vettore tramite la sua applicazione, ossia svolgendo il prodotto A*x. <br> <br>
    

* La matrice speciale che ha la diagonale con tutti 1  chiamata la matrice identit I. <br> <br>
* In generale una matrice $n x m$  una matrice con $n$ righe e $m$ colonne! Se $m=n$ allora la matrice si dice quadra!<br> <br>
* Gli elementi per cui n=m costituiscono la diagonale della matrice.  <br> <br>
Se pensiamo che questa sia la prima volta che vediamo le matrici ci sbaglaimo. Infatti i sistemi lineari non erano nient'altro che il risultato della moltiplicazione tra una matrice e un vettore.

## Metodo di Laplace

<font size="4"> Esiste un altro metodo per calcolare il determinante, il metodo di Lapalce. In formule il teorema di Laplace ci dice che:   <br> <br> 
    $det(A)=\sum(a_{i,j}*(-1)^{i+j}*det(A_{i,j}))$<br> <br> 
A parole  difficile da spiegare, ma ci proviamo. 
* Praticamente si prende un elemento di una riga o colonna di una matrice $(a_{i,j})$ _in maniera intelligente,_ <br> <br> 
    
* eliminiamo dalla matrice la riga iesima e la colonna jesima,<br> <br> 
* calcolaimo il determinante di quello che rimane $det(A_{i,j})$. <br> <br> 
* Ripetiamo questo proccedimento per tutti gli elemnti appartenenti alla riga iesima o alla colonna jesima. <br> <br> Vediamo un esempio:
 $A=\begin{pmatrix} 0 & 0 & 1\\ 1 & 2 & -1 \\ 1 & 2 & -2 \end{pmatrix}$<br> <br> <br> <br>

## Matrice inversa:


<font size="4">Ora che abbiamo introdotto il determinante, possiamo introdurre le matrici inverse. Infatti una matrice  invertibile solamente se ha un determinante diverso da 0!<br> <br>
La matrice inversa di $A$, $A^{-1}$, se esiste,  una matrice tale che: <br> <br>$AA^{-1}=I$<br> <br>
    Vediamo come si calcola la matrice inversa di $A$ :
    
* verifichiamo che  il determinante di $A$ non  zero!, se fosse 0, l'inversa non esisterebbe! <br> <br>
* sostituire ad ogni elemento della matrice il suo cofattore(complemento algembrico.<br> <br>
* calcolare la trasposta della matrice dei cofattori<br> <br>
* dividere la matrice per il determinante.<br> <br>
