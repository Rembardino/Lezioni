# Logaritmi

<font size="4"> Il logaritmo è la funzione inversa dell'esponenziale. <br> <br>
In maniera complicata: il logaritmo di un numero $x$ è l'esponente a cui un altro numero fisso(la base del logaritmo) deve essere elevato per ottenere c.... Proviamo a vederlo in formule:

<font size="5">$log_a(x)=c$ ------> $x=a^c$ </font>
    <br><br>
Quindi prendo la base del logaritmo e la elevo al secondo membre dell'uguaglianza e ottengo il valore per $x$.
<br><br>
    
<font size="6">Esempio</font>:

<font size="5">$log_5(x)=3$ ------> $x=5^3=125$ </font> <br>
<font size="5">$log_3(x)=4$ ------> $x=3^4=81$ </font>
    
<font size="4">Solitamente (controllate!) se la base del logaritmo non è scritta direttamente, si sottointende che la base è 10

<font size="5">$log_{10}(x)=log(x)$
    <br><br>
<font size="4">Esiste anche una base "speciale": il numero di Nepero $e$. In questo caso anzichè scrivere $log_{e}(x)$ si scrive $ln(x)$ per convenzione. <br>

Il numero di Nepero è uguale a $e= 2,718281828459....$ ed è un numero importantissimo. Nepero è lo scopritore dei logaritmmi.


#  Proprietá importanti dei logaritmi

* <font size="5">Il logaritmo è definito <b>solamente</b> per $x$>0!!!<br> <br>
* Regola del prodotto: <font size="5">$log_{b}(x*y)=log_{b}(x)+log_{b}(y)$
    <br><br>
* Regola della divisione <font size="5">$log_{b}(\frac{x}{y})=log_{b}(x)-log_{b}(y)$ <br> <br>
* Regola della potenza <font size="5">$log_{b}(x^q)=q*log_{b}(x)$ <br> <br>
* Regola della radice <font size="5">$log_{b}(\sqrt[p]x)=\frac{log_{b}(x)}{p}$
    <br><br>
    a:
* Cambio di base ${\displaystyle \log _{b}x={\frac {\log _{k}x}{\log _{k}b}}.\,}{\displaystyle \log _{b}x={\frac {\log _{k}x}{\log _{k}b}}.\,}$

<font size="4"> Il logaritmo è la funzione inversa dell'esponenziale. <br> <br>
        In maniera complicata: il logaritmo di un numero $x$ è l'esponente a cui un altro numero fisso(la base del logaritmo) deve essere elevato per ottenere c....<font size="5">$log_a(x)=c$ ------> $x=a^c$ </font>
    

    
    

### Vediamo come lo stesso ragionamento si applica alle altre operazioni più familiari: <br><br>
<font size="4">
 La sottrazione è la funzione inversa della somma: $8-x=2$<br>
 $x$ e' il numero che sommato a 2 mi da 8 <br><br>
 La divisione è la funzione inversa della moltiplicazione: $8:x=2$<br>
 $x$ e' il numero che moltiplicato per 2 mi da 8 <br><br>
 La radice quadrata è la funzione inversa dell'elevamento a potenza: $\sqrt{16}=x$<br>
 $x$ è il numero che elevato al quadrato (cubo etc...) mi da 16.<br><br>
 
Fino ad ora, abbiamo sempre conosciuto esplicitamente l'esponente delle nostre potenze: <br><br>2 per  $x^2$ <br><br> 3 per  $x^3$ <br><br> oppure <br><br> $\frac{1}{2}$ per $\sqrt{x}$ <br><br> $\frac{1}{3}$ per $\sqrt[3]{x}$

 

### Funzioni Esponenziali <br>
    
 <font size="4">
Da ora in poi permettiamo anche all'esponente della potenza di essere un'incognita $x$. In questo modo creiamo le funzioni esponenziali: <font size="5"> $5^x$, $8^x$ </font> ecc... <br><br>
$5^x$ vuol dire che la <b>base</b> 5 è moltiplicata per se stessa $x$ volte. $x$ è l'esponente. Vediamo graficamente la funzione esponenziale in diversi casi

```python
#esponenziale
#parametri per grafici ed altro non modificare####

import matplotlib as mpl, matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(-1.5,1.5,0.005) #!non modificare!#

y1=5**x


#!non modificare!#
plt.figure()
plt.grid()
plt.plot(x,y1,label='5^x',color='green')
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(0,1,s = 100, color='red',label="y=1")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.legend()
plt.show()
```

<font size="4"> Vediamo alcune proprietá interessanti di questo grafico: 
* Vediamo che per $x=0$ il grafico passa per $y=1$
* Per quanto $x$ diminuisca, sembra che la funzione non arrivi mai al valore $y=0$
* Per valori grandi di $x$ la funzione cresce molto velocemente

## Proviamo ora a vedere come cambia la curva cambiano il valore della base
<font size="4"> Usiamo come base 2, il numero di Nepero e (=2.718...), 5 e 10.

```python
#esponenziale
#parametri per grafici ed altro non modificare####

import matplotlib as mpl, matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(-1.5,1.5,0.005) #!non modificare!#

y=5**(x)
y1=np.exp(x) #logaritmo in base e
y2=10**(x) #base 10
y3=2**(x) #base 2



#!non modificare!#
plt.figure()
plt.grid()
plt.plot(x,y3,label='2^x')
plt.plot(x,y1,label='e^x')
plt.plot(x,y,label='5^x')
plt.plot(x,y2,label='10^x')
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(0,1,s = 100, color='red',label="y=0")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.legend()
plt.show()
```

<font size="4"> Vediamo come all'aumentare della base, la crescita si fa molto piu ripida. Infatti se poniamo $x=3$ vediamo che $2^3=8$,  $5^3=125$, $10^3=1000$. <br> <br>
Ora vediamo come si comporta in dettaglio la funzione $2^x$:<br> <br> $2^1=2$, $2^2=4$, $2^3=8$, $2^4=16$, $2^5=32$....<br> <br>vediamo che cresce molto velocemente. Questa crescita è detta appunto crescita esponenziale!. <br> <br>
 

# Proprietà della funzione esponenziale:
<font size="4"> Le proprietà della funzione esponenziale sono le stesse delle potenze:
* $a^m*a^n=a^{m+n}$ <br><br>
* $\frac{a^m}{a^n}=a^{m-n}$<br><br>
* $(a^m)^n=a^{m*n}$<br><br>
* $a^s*b^s=(a*b)^{s}$<br><br>
* $a^{-m}=\frac{1}{a^m}$<br><br>

```python
#logaritmo
#parametri per grafici ed altro non modificare####

import matplotlib as mpl, matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0.005,10,0.005) #!non modificare!#

y1=np.log(x) #logaritmo in base e
y2=np.log10(x) #base 10
y3=np.log2(x) #base 2



#!non modificare!#
plt.figure()
plt.grid()
plt.plot(x,y3,label='base 2')
plt.plot(x,y1,label='base e')
plt.plot(x,y2,label='base 10')
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(1,0,s = 100, color='red',label="x=0")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.legend()
plt.show()
```
