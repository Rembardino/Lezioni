<font size="8"> <b> I 3 principi della dinamica</b>

 <font size="5"> I 3 principi della dinamica vennero presentati nel 1687 da Isacc Newton. Questi principi vengono anche spesso chiamati le tre leggi di Newton. <br><br>
<b>Primo principio della dinamica:</b> _Se su di un corpo non sono applicate forze, oppure se la risultante delle forze applicate su di esso è nulla,    allora il corpo resta fermo se è in uno stato di quiete. Oppure, se si st muovendo, continuerà a muoversi di <font color='red'>moto rettilineo uniforme. </font>_  <br><br>
<b> Secondo principio della dinamica:</b> _Se viene applicata una forza a un ogetto, quest'ultimo subirà un'<font color='darkcyan'>accellerazione </font> nello stesso <font color='darkorange'>verso </font> e <font color='green'>direzione</font> della forza applicata. Allo stesso modo, se la risultante delle forze applicate dal corpo è diversa da zero, allora il corpo subirà un' <font>accellerazione </font> nello stesso <font color='darkorange'>verso </font> e <font color='green'>direzione</font> della risultante delle forze_ <br><br>
<b>Terzo principio della dinamica:</b> _Se un corpo $A$ applica una forza su un corpo $B$, allora il corpo $B$ applica una forza con la stessa <font color='green'>direzione</font>, <font color='magenta'>intensitá</font> ma <b><font color='darkorange'>verso opposto</font></b>  su di $A$._

<font size="6"><b>Grandezze vettoriali e grandezze scalari </b>

<font size="4">In fisica possiamo definire due tipi di grandezze: grandezze _scalari_ e grandezze _vettoriali_. <br><br>
    Mentre le grandezze scalari hanno bisogno del solo valore per descriverle completamente, le grandezze vettoriali hanno bisogno anche di altre informazioni per descriverle completamente.

<b>Esempi:</b><br><br>
1) _Quanti gradi ci sono?_ Ci sono 25°C<br><br>
Non ho bisogno di altre informazioni per capire quanti gradi ci sono. Questo perchè la Temperatura è una grandezza <b>scalare</b>.<br><br><br>
2) Se leggiamo su una porta leggiamo: _Per aprire la porta applicare una forza di almeno 1 N_.<br><br>
In questo caso sappiamo quanta forza dobbiamo applicare alla porta per aprirla, ma non sappiamo ancora <b><font color='blue'>dove</font></b> dover applicare la forza, in che   <b><font color='green'>direzione</font></b> dobbiamo applicarla e se dobbiamo <b><font color='darkorange'> spingere o tirare</font></b>. Questo perchè la Forza è una grandezza <b>vettoriale</b>!<br><br><br>
    
Quindi, per descrivere completamente una grandezza vettoriale abbiamo bisogno del <b><font color='blue'>punto di applicazione</font></b> della forza, della sua <b><font color='green'>direzione</font> </b>, del suo  <b><font color='darkorange'>verso</font></b> e del suo  <b><font color='magenta'>modulo</font></b>.<br> <br><br>
Esempi di grandezze scalari sono:
* il tempo [secondi,s]<br><br>
* la massa [chilogrammi, Kg]<br><br>
* la temperatura [gradi Kelvin,K]<br><br>
<br> 

Esempi di grandezze vettoriali sono:
* La forza [Newton,N]<br><br>
* la velocita [metri al secondo,$\frac{m}{s}$]<br><br>
* L'accellerazione [metri a secondo quadrati, $\frac{m}{s^2}$]<br><br>


Per distinguere tra grandezze scalari e vettoriali si utilizza spesso scrivere una freccia sopra le grandezze vettoriali $\stackrel{\rightarrow}{F}$ o scriverle semplicemente in grassetto <b>F</b>

Anche se questo sembra una complicazione, possiamo trarre vantaggio dal fatto che le forze sono grandezze vettoriali. Per prima cosa possiamo visualizzare una forza diseganndo una freccia che rappresenta interamente la nostra forza.

<font size="6"><b>Secondo principio della dinamica</b><br><br>
<font size="5">_Se viene applicata una forza a un ogetto, quest'ultimo subirà un' <font color='darkcyan'>accellerazione </font> nello stesso verso e direzione della forza applicata. Allo stesso modo, se la risultante delle forze applicate dal corpo è diversa da zero, allora il corpo subirà un' <font>accellerazione </font> nello stesso verso e direzione della risultante delle forze._ <br>
    
La prima cosa che ci dice questo principio è che per <font color='darkcyan'>accellerare </font> un corpo dobbiamo applicare una forza!.<br>
<font color='blue'>ATTENZIONE! per accellerazione si intende un qualsiasi cambiamento della velocità! Quindi anche una "decellerazione" è sempre una accellerazione ma con verso opposto al moto del punto, quindi diminuisce la velocità!!!!! </font>   
<br>Newton si accorse che esiste una relazione tra la forza applicata al corpo e l'accellerzione impressa. Aumentando la forza, aumentava anche l'accellerazione. Tuttavia si accorse che la stessa forza non accellerava allo stesso modo corpi con diversa massa. Allora lui scrisse la seguente formula che rappresentava benissimo la relazione di dipendenza tra la forza applicata, la massa del corpo e l'accellerazione impressa al corpo nella seguente famosissima formula:<br><br>
<b><font size="8">
    $\stackrel{\rightarrow}{F}=m\stackrel{\rightarrow}{a}$</font></b>

<br>
Questa formula ci dice che se aumentiamo la forza applicata ad un corpo di massa $m$ anche la sua accellerazione aumenta. Se diminuiamo l'intensità della forza applicata al corpo vediamo che anche l'accellerazione impressa diminusice. In maniera formale si dice che forza e accellerazione sono <b>direttamente proporzionali</b>.
    
Ora possiamo capire perchè il corpo non soggetto a forze (o se <b>R</b>=0) in movimento si muove di moto rettilineo uniforme!. Non essondoci nessuna forza, non può esserci nessuna variazione di velocità. $a=0$ quindi $v=costante$
    
    
    

 <div>

</div>
<br><br>

<font size="5">Questa formula ci aiuta anche a ricordare come è definita l'unità di misura della forza, ossia il Newton. Abbiamo detto che la massa si misura in Kg, l'accellerazione in $\frac{m}{s^2}$ e la forza in Newton.Ma allora: <br><br>

   <font size="6"> _$N=Kg*\Large\frac{m}{s^2}$_ </font>
 <br><br>
       
<font color='blue'>L'analisi dimensionale è fondamentale per capire se un esercizio è venuto bene o no e vi da anche dei suggerimenti su quali formule usare. <br>
<br>Esempio:<br> Il problema vi da un tempo in secondi e una distanza in metri e vi chiede la velocità. Mettiamo caso non vi ricordate la formula di come si calcola la velocità. Quello che però sapete è che la velocità si misura in $\frac{m}{s}$. <br> <br> 
    Allora capite subito che dovete dividere la distanza [m] per il tempo [s]. Ed ecco qui che avete fatto bene un esercizio. Inoltre se le unitá di misura tornano, potete sapere subito che avete fatto l'esercizio bene (con una buona probabilità). <font><br> <br> 

<font size="6"><b>Terzo principio della dinamica</b>

<font size="4">Se un corpo $A$ applica una forza su un corpo $B$, allora il corpo $B$ applica una forza con la stessa <font color='red'>direzione</font>, <font color='red'>intensitá</font> ma <font color='red'>verso opposto</font>  su di $A$.
    
Questo principio è spesso chiamato come il principio di _"azione e reazione"._ Questo principio è rispettato ogni volta che camminiamo: il nostro piede spinge il pavimento verso il basso mentre il pavimento spinge il nostro piede verso l'alto. Quando spingiamo un oggetto, esso tende a spingerci anche a noi. La forza di _"reazione"_ ha lo stesso <b><font color='magenta'>modulo</font></b>, stessa <b><font color='green'>direzione</font></b> ma <b><font color='darkorange'>verso opposto!</font></b>. Possiamo scrivere questo concetto in formula:
<font size="8"><br><br> <b>$\stackrel{\rightarrow}{F}_{AB}=-\stackrel{\rightarrow}{F}_{BA}$</font></b> 
<br><br>
Succede la stessa cosa quando appoggiamo un corpo su di un tavolo: Il  corpo applica una forza al tavolo (La Forza peso) e il tavolo applica una forza con lo stesso modulo e direzione ma veso opposto (reazione vincolare). E visto che la risultante delle due forze è nulla, per il primo principio, il corpo rimane fermo sul tavolo.

<font size="4"> Abbiamo un piano inclinato e un corpo sopra di esso come in figura:

<font size="4">Visto che siamo sulla Terra il corpo è soggetto alla Forza di gravitá rivolta verso il basso <font color='red'><b>$\stackrel{\rightarrow}{F}_{P}$</b></font>. Abbiamo visto tutti un corpo scivolare su di un piano inclinato. Ma le leggi di Newton ci dicono che il corpo dovrebbe muoversi nella direzione e verso della forza applicata. In questo caso _sembra_ non sia cosi.<br> <br>
Proviamo ad analizzare meglio la situazione. Possiamo provare a _<b>scomporre</b>_ la forza peso senza alterare minimamente il sistema.


<div>

<font size="4">Quello che abbiamo fatto è immaginare la <font color='red'><b>$\stackrel{\rightarrow}{F}_{P}$</b></font> come la risultante di due altre forze. Queste forze vengono rispetttivamente chiamate forza paralla <font color='red'><b>$\stackrel{\rightarrow}{F}_{//}$</b></font> e forza perpendicolare <font color='red'><b>$\stackrel{\rightarrow}{F}_{⟂}$</b></font>. Non Abbiamo fatto nulla, sappiamo che quello che conta è la risultante delle forze! <br> <br>
    Per rendere (spero) più chiare le figure, e visto che abbiamo scomposto la forza, possiamo togliere completamente la freccia rossa che descrive la forza peso. Quello che facciamo è sostituirla con una freccia un pochino meno "invasiva". Nella figura sottostante ho messo un riquadro per capire meglio la composizione geometrica delle forze. Questo riquadro ci aiuterá anche tra poco per capire una cosa, ma non faccio spoiler. 
    <div>

    </div>

<font size="4">Ora vediamo che <font color='red'><b>$\stackrel{\rightarrow}{F}_{⟂}$</b></font> è una forza applicata al nostro piano inclinato. Ma Per il terzo principio della dinamica, il piano applica una forza dello stesso modulo e stessa direzione ma verso opposto! Questa forza si chiama reazione vincolare oppure forza "normale" <font color='green'><b>$\stackrel{\rightarrow}{N}$</b></font>

<div>

    </div>

<font size="4"> Quindi possiamo cancellare completamente la componente della forza perpendicolare con la reazione vincolare del piano visto che sono due forze con stesso modulo e direzione ma verso opposto quindi la loro risultante è 0! <br> <br> 

<div>

    </div>

<font size="4">Quello che ora rimane è la componente della forza parallela al piano. <b>Quindi i principi (non i conti) tornano</b>. Il corpo si muove nella direzione e verso della risultante delle forze! <br> <br>
