# Trovare il dominio della funzione

<font size="4"> Il dominio della funzione sono tutti i punti per cui la nostra funzione è definita. Vediamo di capirlo con qualche esempio: <br> <br>
$f(x)=x$


```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)

x=np.arange(-1,1,0.005) #!non modificare!#


y=x #!non modificare!#

plt.figure()
plt.grid(linewidth=2)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

<font size="4">Vediamo come la funzione è definita in tutto R, quindi il dominio è da $-\infty$ a $\infty$. Ma cosa succede se abbiamo una x al denominatore: per esempio la funzione <br> <br>
$f(x)=\frac{1}{x}$

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)

x=np.arange(-2,2,0.003) #!non modificare!#


y=1/x #!non modificare!#

plt.figure()
plt.grid(linewidth=2)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-2,2)
plt.ylim(-20,20)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

<font size="4"> Vediamo che la funzione si avvicina sempre di più al valore x=0 senza mai arrivarci. Questo perchè non si può dividere per 0! Quindi in questo caso il dominio della funzione è tutto R <b>tranne  </b>lo 0!. Allo stesso modo, se avessimo avuto la funzione: <br> <br>
$f(x)=\frac{1}{(x-1)(x+2)(x-80)}$ <br> <br>
dovremmo escludere tutti i valori di x per cui il denominatore diventa 0. Quindi in questo caso x=1,x=-2 e x=80. Per questi valori di x infatti la funzione non esisterebbe perchè avremmo uno 0 al denominatore.

<font size="4"> Se invecec avessimo la funzione:  <br> <br>
    $f(x)=\sqrt{x}$ <br> <br>
Il dominio sarebbe $x\ge0$ perchè l'argomento della radice non può essere negativo. Può essere 0 ma non minore di 0

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)

x=np.arange(0,2,0.003) 


y=x**(1/2)

plt.figure()
plt.grid(linewidth=2)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-0.5,2)
plt.ylim(-5,5)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

<font size="4"> Ultimo esempio importante è la funzione logaritmo:<br><br>
    $f(x)=log(x)$
In lezioni prececdenti abbiamo visto in dettaglio la funzione logaritmo. Dovreste quindi ora sapere che il logaritmo è definito solamente per x>0! <br><br>
Ovviamente può capitare che abbiamo una funzione complicata come:<br><br>
    $f(x)=\frac{log(x)}{(x-1)\sqrt{x+4}+log(x-2)}$<br><br>
In questo caso dobbiamo tirarci su le maniche e vedere pezzo per pezzo tutte le varie condizioni da verificare:
* Al numeratore abbiamo log(x) quindi x > 0
* Al denominatore abbiamo $x-1\neq0$, quindi $x\neq1$
* Al denominatore abbiamo anche $\sqrt{x+4}$ quindi x+4>0 quindi x>-4 
* abbiamo anche log(x-2) quindi $x-2\neq0$ quindi $x\neq2$,
* infine abbiamo anche che ${(x-1)\sqrt{x+4}+log(x-2)}\neq0$<br><br>
Il dominio della funzione è l'unione di tutti i domini delle varie sub-funzioni.
    

# Capire se la funzione è pari o dispari


<font size="4"> Perchè è importante questo step?
Bene se dovessimo trovare che la funzione da studiare è pari o dispari, questo ci permetterebbe di semplificare di molto i nostri conti. Infatti:<br><br>
* Se una funzione è _pari_ allora  possiamo studiare la nostra funzione solamente per valori positivi di $x$. Infatti l'asse delle $y$ è un asse di simmetria per la nostra funzione. Questo vuol dire che possiamo "ribaltare" il grafico per $x >0$ rispetto all'asse $y$ per avere quello che succede per le $x < 0$. Vediamo qualche esempio 

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)

x=np.arange(-20,20,0.003) 


y=x**2
y1=np.cos(x)

plt.figure()
plt.grid(linewidth=2)
plt.plot(x,y)
plt.plot(x,y1)
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-5,5)
plt.ylim(-2,2)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

<font size="4"> 
* Anche se la funzione è dispari possiamo studiare la nostra funzione solamente per valori positivi di $x$. In questo caso sará l'origine il nostro punto di simmetria. Per capire meglio come "ribaltare" in questo caso i valori ottenuti per $x>0$ per ottenere i valori per $x < 0$ vediamo alcuni esempi:

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)

x=np.arange(-20,20,0.003) 


y=x**3
y1=np.sin(x)

plt.figure()
plt.grid(linewidth=2)
plt.plot(x,y)
plt.plot(x,y1)
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

<font size="4"> Ma come facciamo a capire se una funzione è pari o dispari? Dobbiamo verificare 2 condizioni:
* se $f(x)=f(-x)$ allora la funzione è _pari_ <br><br>
* se $f(-x)=-f(x)$ allora la funzione è _dispari_ <br><br>
Per capire meglio le definizioni facciamo qualche esempio:<br><br>

    

##     Funzioni pari:
* $f(x)=x^2$. Abbiamo che $f(-x)=(-x)^2=x^2$ e quindi $f(x)=f(-x)$ in quanto $x^2=x^2$<br><br>
* $f(x)=cos(x)$. Potremmo utilizzare le regole trigonometriche per dimostrarlo ma sarebbe un pochino ridondante. quindi vediamo che succede per alcuni valori di x. Per esempio prendiamo 45° $(\frac{\pi}{4})$, 90° $(\frac{\pi}{2})$ e -45°  $(-\frac{\pi}{4})$ e -90°$(-\frac{\pi}{2})$ e calcoliamo i valori del coseno per quesi angoli:<br><br>
$cos(\frac{\pi}{4})=\frac{1}{\sqrt{2}}$,$cos(\frac{\pi}{2})=0$<br><br>
$cos(-\frac{\pi}{4})=\frac{1}{\sqrt{2}}$,$cos(-\frac{\pi}{2})=0$   <br><br>
Vediamo quindi come $f(x)=f(-x)$.

##     Funzioni dispari:
* $f(x)=x^3$. Abbiamo che $f(-x)=(-x)^3=-x^3$, $-f(x)=-x^3$ e quindi $f(-x)=-f(x)$ in quanto $-x^3=-x^3$<br><br>
* $f(x)=sen(x)$. Anche qui potremmo utilizzare le regole trigonometriche per dimostrarlo ma sarebbe un pochino ridondante. quindi vediamo che succede per alcuni valori di x (gli stessi di prima). 
Prendiamo 45° $(\frac{\pi}{4})$, 90° $(\frac{\pi}{2})$ e -45°  $(-\frac{\pi}{4})$ e -90°$(-\frac{\pi}{2})$ e calcoliamo i valori del seno per quesi angoli:<br><br>
$sen(-\frac{\pi}{4})=-\frac{1}{\sqrt{2}}$,$sen(-\frac{\pi}{2})=-1$<br><br>
$-sen(\frac{\pi}{4})=-\frac{1}{\sqrt{2}}$,$-sen(\frac{\pi}{2})=-1$   <br><br>
Vediamo quindi come $f(-x)=-f(x)$.

# Trovare le intersezioni con gli assi

<font size="4"> Possiamo ora cominciare a disegnare il nostro grafico. Il punto di inizio piú facile è quello di cercare di capire dove la nostra funzione interseca i due assi (x e y) 

<font size="4"> Come spiegato in una lezione passata, l'intersezione degli assi si ottiene ponendo x o y =0. Per trovare quando la funzione interseca l'asse delle y, dobbiamo uguagliare l'intera funzione a 0. Infatti ci chiediamo quando $f(x)=0$.<br><br> Per esempio se $f(x)= x^2+4x+3$ dobbiamo risolvere $x^2+4x+3=0$. Risolvendo troviamo che le soluzioni di questa equazione sono -3 e -1. Questi allora saranno i valori dove la funzione interseca l'asse delle x.<br><br> Se vogliamo trovare invece i valori della funzione dove interseca l'asse delle y allora dobbiamo porre la nostra x=0. Infatti ricordiamoci che $f(x)$ non è nient'altro che il valore y della funzione al valore x. Infattti possiamo scrivere $f(x)=y$. Quindi poniamo x=0 e abbiamo $f(x=0)=y$. Nell'esempio precedente abbiamo che $0^2+4*0+3=y$ quindi $y=3$. La funzione interseca l'asse delle y al valore y=3

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)

x=np.arange(-8,10,0.003) 


y=x**2+4*x+3

plt.figure()
plt.grid(linewidth=2)
plt.plot(x,y)

plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-5,5)
plt.ylim(-4,4)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

<font size="4"> Ovviamente questo è  un caso semplicissimo. Se dovessimo avere funzioni più complicate allora la soluzione potrebbe essere diversa. <br> <br> Per esempio, se abbiamo funzioni periodiche, ci saranno molto punti dove la funzione interseca l'asse delle x.

# Limiti nei punti "critici" nel dominio

<font size="4"> Ora dobbiamo chiederci cosa succcede vicino ai punti "critici" della nostra funzione. I punti critici sono i punti per i quali la funzione non è definita. Ok, in quei punti precisi non è definita la funzione, ma cosa succede vicinissimo quei punti? E all'infinito? <br><br> 
Per rispondere a queste domande dobbiamo calcolare il limite della funzione rispetto a questi punti. Se dobbiamo analizzare il comportamento della funzione vicicno un asintoto verticale, dobbiamo studiare la funzione a sinistra e a destra del punto come mostrato in figura. 

   <div>

</div>

<font size="4"> Questo si fa faccendo il limite sinitro e il limite destro della nostra funzione. Cerchiamo di capire cosa vuol dire questa cosa con la funzione $\frac{1}{x}$. Sappiamo che per x=0 la funzione non è definita, ma cosa succede poco più a destra di 0 ($0^+$) o poco piu a sinistra ($0^-$) 

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)

x=np.arange(-2,2,0.003) #!non modificare!#


y=1/x #!non modificare!#

plt.figure()
plt.grid(linewidth=2)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-2,2)
plt.ylim(-20,20)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

<font size="4"> Se facciamo il limite da destra abbiamo che $f(x)=\frac{1}{0^+}$. Che valore ha qui la funzione? positivo o negativo?. Possiamo pensare che $0^+$ è positivo allora abbiamo che $\frac{1}{0^+}$ tende a   $+\infty$. Viceversa, "venendo" da sinistra, $\frac{1}{0^-}$ è negativo quindi $\frac{1}{0^-}$ tende a $-\infty$

<font size="4"> Vediamo un'altra funzione: $f(x)=\frac{1}{(x+2)^2}$. Studiamo il limite da destra <br> <br>
    $\lim_{x \to -2^+}\frac{1}{(x+2)^2}$ Questo diventa: $f(x)=\frac{1}{(0^+)^2}$.<br> <br>
    $0^+$ è un  numero piccolissimo positivo che elevato al quadrato è ancora positivo quindi la funzione tende a $+\infty$ <br> <br>
     $\lim_{x \to -2^-}\frac{1}{(x+2)^2}$ diventa $f(x)=\frac{1}{(0^-)^2}$. <br> <br>
    $0^-$ è un  numero piccolissimo negativo che elevato al quadrato è diventa positivo quindi la funzione tende a $+\infty$. Vediamo ora il grafico di $f(x)=\frac{1}{(x+2)^2}$

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)

x=np.arange(-4,4,0.003) #!non modificare!#


y=1/((x+2)**2) #!non modificare!#

plt.figure()
plt.grid(linewidth=2)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-4,0)
plt.ylim(-5,50)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

<font size="4"> Vediamo come quindi la funzione tende a $+\infty$ sia a destra che a sinistra.

# Studio della derivata prima

<font size="4"> Ora possiamo cominciare ad ottenere informazioni più dettagliate riguardo la nostra funzione. Possiamo calcolare i punti di massimo e minimo ponendo la derivata prima della nostra funzione uguale a 0. Ci sarebbe molto da dire sulla derivata e sul suo significato. Sicuramente farò una lezione su questo argomento prima o poi. Per adesso ricordo che la derivata prima ci dice se la funzione sta crescendo se la derivata è > 0 ($f'(x) >0$  o se la funzione sta decrescendo se la derivata è < di 0 $f'(x) < 0 $. <br> <br>
    
 <div>

</div>
    
<br><br>
Quindi una volta trovati i punti di massimi o minimi sappiamo dove la funzione cresce o decresce. Quello che non sappiamo è la concavitá della funzione. Per questo serve la derivata seconda.


# Studio della derivata seconda

<font size="4"> La derivata seconda ci dice la concavitá della funzione. Se la derivata seconda è > 0 ($f''(x) > 0 $ la concavità  della funzione è rivolta verso l'alto, altrimenti se la derivata seconda è <0 ($f''(x) < 0$) la conccacvità è rivolta verso il basso.
Cosa succede quando la derivata seconda è uguale a 0? ($f''(x) = 0$) <br> <br>


Questi punti si chiamano _punti di flesso_. Questi punti non sono ne di massimo ne di minimo ma sono soloamente punti dove la funzione cambia concavità.
