# Logaritmi

<font size="4"> Il logaritmo è la funzione inversa dell'esponenziale. <br> <br>
In maniera complicata: il logaritmo di un numero è l'esponente a cui un altro numero fisso (la base del logaritmo) deve essere elevato per ottenere $x$.... Proviamo a vederlo in formule:

<font size="5">$log_a(x)=c$ ------> $x=a^c$ </font>
    <br><br>
Quindi prendo la base del logaritmo e la elevo al secondo membre dell'uguaglianza e ottengo il valore per $x$.
<br><br>
    
Esiste anche una base "speciale": il numero di Nepero $e$. In questo caso anzichè scrivere $log_{e}(x)$ si scrive $ln(x)$ per convenzione. <br>

Il numero di Nepero è uguale a $e= 2,718281828459....$ ed è un numero importantissimo. Nepero è lo scopritore dei logaritmmi.


#  Proprietá importanti dei logaritmi

* <font size="5">Il logaritmo è definito <b>solamente</b> per $x>0$!!!<br> <br>
* Regola del prodotto: <font size="5">$log_{b}(x*y)=log_{b}(x)+log_{b}(y)$
    <br><br>
* Regola della divisione <font size="5">$log_{b}(\frac{x}{y})=log_{b}(x)-log_{b}(y)$ <br> <br>
* Regola della potenza <font size="5">$log_{b}(x^q)=q*log_{b}(x)$ <br> <br>
* Regola della radice <font size="5">$log_{b}(\sqrt[p]x)=\frac{log_{b}(x)}{p}$
    <br><br>
    a:
* Cambio di base ${\displaystyle \log _{b}x={\frac {\log _{k}x}{\log _{k}b}}.\,}{\displaystyle \log _{b}x={\frac {\log _{k}x}{\log _{k}b}}.\,}$

### Vediamo come lo stesso ragionamento si applica alle altre operazioni più familiari: <br><br>
<font size="4">
 La sottrazione è la funzione inversa della somma: $8-x=2$<br>
 $x$ e' il numero che sommato a 2 mi da 8 <br><br>
 La divisione è la funzione inversa della moltiplicazione: $8:x=2$<br>
 $x$ e' il numero che moltiplicato per 2 mi da 8 <br><br>
 La radice quadrata è la funzione inversa dell'elevamento al quadrato: $\sqrt{16}=x$<br>
 $x$ è il numero che elevato al quadrato (cubo etc...) mi da 16.<br><br>
 
Fino ad ora, abbiamo sempre conosciuto esplicitamente l'esponente delle nostre potenze: <br><br>2 per  $x^2$ <br><br> 3 per  $x^3$ <br><br> oppure <br><br> $\frac{1}{2}$ per $\sqrt{x}$ <br><br> $\frac{1}{3}$ per $\sqrt[3]{x}$

# Funzioni Esponenziali <br>
    
 <font size="4">
Da ora in poi permettiamo anche all'esponente della potenza di essere un'incognita $x$. In questo modo creiamo le funzioni esponenziali: <font size="5"> $5^x$, $8^x$ </font> ecc... <br><br>
$5^x$ vuol dire che la <b>base</b> 5 è moltiplicata per se stessa $x$ volte. $x$ è chiamato esponente. Vediamo graficamente la funzione esponenziale $5^x$

```python
#esponenziale 5^x


import matplotlib as mpl, matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(-1.5,1.5,0.005) 
#x=np.arange(-10.5,-5.5,0.005) 

y1=5**x


#!non modificare!#
plt.figure()
plt.grid()
plt.plot(x,y1,label='5^x',color='green')
plt.title("5^x")
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(0,1,s = 100, color='red',label="y=1")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.legend()
plt.show()
```

<font size="4"> Vediamo alcune proprietà interessanti di questo grafico: 
* Vediamo che per $x=0$ il grafico passa per $y=1$<br><br>
* Per quanto $x$ diminuisca, sembra che la funzione $5^{x}$ non arrivi <b>mai</b> al valore $y=0$. In realtà ci arriva, ma all'inifinito: $5^{-\infty}=\frac{1}{5^{\infty}}=0$<br><br>
* Per valori grandi di $x$ la funzione cresce molto velocemente<br><br>

## Proviamo ora a vedere come cambia la curva cambiando il valore della base
<font size="4"> Usiamo come base dell'esponenziale 2, il numero di Nepero (=2.718...), 5 e 10.

```python
#esponenziale
#parametri per grafici ed altro non modificare####

import matplotlib as mpl, matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(-1.5,1.5,0.005) #!non modificare!#

y=5**(x)
y1=np.exp(x) #logaritmo in base e
y2=10**(x) #base 10
y3=2**(x) #base 2



#!non modificare!#
plt.figure()
plt.grid()
plt.title("Esponenziali con diverse basi")
plt.plot(x,y3,label='2^x')
plt.plot(x,y1,label='e^x')
plt.plot(x,y,label='5^x')
plt.plot(x,y2,label='10^x')
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(0,1,s = 100, color='red',label="y=0")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.legend()
plt.show()
```

<font size="5"> Vediamo come all'aumentare della base, la crescita (per x>0) si fa molto piu ripida. Infatti se poniamo $x=3$ vediamo che $2^3=8$,  $5^3=125$, $10^3=1000$. <br> <br>
Ora vediamo come si comporta in dettaglio la funzione $2^x$:<br> <br> $2^0=1$, $2^1=2$, $2^2=4$, $2^3=8$, $2^4=16$, $2^5=32$....<br> <br>vediamo che il valore cresce molto velocemente. Questa crescita è detta appunto <font color="red"> crescita esponenziale!</font>.<br><br> In questo periodo pandemico abbiamo sicuramente sentito le parole _crescita_ e _esponenziale_ insieme. Ora capiamo perchè la crescita esponenziale dei contagi è un problema. Il numero aumenta velocemente!<br> <br>
Dal grafico delle diverse funzioni vediamo anche che 
* per $x=0$ tutte le funzioni hanno il valore $y=1$<br><br>
* per quanto $x$ diminuisca,tuttle le funzioni sembrano non arrivare mai al valore $x=0$. Come detto prima, in realtà ci arrivano, ma a $x=-\infty$!<br>

# Proprietà della funzione esponenziale:
<font size="5"> Le proprietà della funzione esponenziale sono le stesse delle potenze:<br><br>
    
    
<h1><center> $a^m*a^n=a^{m+n}$       $\;\;\;\;\;\;$$\;\;\;\;\;\;$    $5^4*5^3=5^{4+3}=5^{7}$ <br><br>
 $\frac{a^m}{a^n}=a^{m-n}$   $\;\;\;\;\;\;$$\;\;\;\;\;\;$   $\;\;\;\;$       $\frac{6^5}{6^2}=6^{5-2}=6^{3}$ <br><br>
 $(a^m)^n=a^{m*n}$   $\;\;\;\;\;\;$$\;\;\;\;\;\;$     $\;\;$             $(3^4)^4=3^{4*4}=3^{16}$ <br><br>
 $a^s*b^s=(a*b)^{s}$    $\;\;\;\;\;\;$$\;\;\;$               $3^5*2^5=(3*2)^{5}=(6)^{5}$ <br><br>
 $a^{-m}=\frac{1}{a^m}$  $\;\;\;\;\;\;$$\;\;\;\;\;\;$  $\;\;\;\;$    $\;$       $5^{-2}=\frac{1}{5^2}$ <br><br></center></h1>


<font size="4"> Ora cerchiamo di capire meglio il fatto che qualsiasi esponenziale è zero solamente a $x={-\infty}$. <br><br>Prendiamo per esempio $5^x$ e poniamo $x=-\infty$, otteniamo quindi $5^{-\infty}=0$. Per le proprietà delle potenze possiamo riscrivere $5^{-\infty}=\frac{1}{5^{\infty}}$. <br><br>
$5^{\infty}$ è un numero grandissimo ora e quindi dovrebbe essere più facile capire che dividere un numero per qualcosa di _infinitamente_ grande sia un numero piccolissimo.

<font size="4"> I logaritmi sono spesso un tallone d'Achille per gli studenti. Il solo nominarli a volte provoca il vomito. A mio avviso questo è dovuto al fatto che non ne capiscono il significato e l'uso!<br> <br>
I Logaritmi vennero introdotti da Neplero nel 1614. Inizialmente la connessione tra logaritmi e funzioni esponenziali non era notata. I logaritmi venivano utilizzati solamente perchè <b> semplificavano</b> i conti. Infatti, come vedremo a breve, i logaritmi permettono di trasformare le moltiplicazioni in semplici addizioni tramite la formula: 
  <h2><center><font size="5">  $log_{t}(a*b)=log_{t}(a)+log_{t}(b)$</center></h2>

 Solamente nel 1730 Eulero collegò esplicitamente i logaritmi con le funzioni esponenziali tramite la formula  <h1><center> $x=log_{t}(a)$</center></h1>
  Dove $x$ è l'esponente a cui bisogna elevare t per ottenere il numero a: <h1><center>$a=t^x$</center></h1>
Proviamo ora a vedere visivamente il grafico del logaritmo. Per esempio vediamo il grafico di  $log_{10}(x)$

Quando è che questa ugualianza è vera? Vediamo che la base dei logaritmi è la stessa, ma allora, affinchè l'uguaglianza sia vera, gli arogmenti dei logaritmi devono essere uguali! Quindi: <br><br>
    $\frac{x}{7}=(x-1)$  Ora abbiamo una semplice equazione di primo grado: <br> <br>
    $x=7x-7$<br><br>
    $-6x=-7$<br><br>
    $6x=7$<br><br>
    $x=\frac{7}{6}$<br><br>
    $\frac{7}{6}$ è maggiore di 1 e quindi stiamo apposto!
