# Parabole

La tipica formula di una parabola Ã¨  

<font size="4">$y=ax^2+bx+c$</font>

dove $a,b$ e $c$ sono numeri positivi o negativi. Il valore e il segno di questi tre parametri definiscono la posizione della parabola nello spazio. 

_Nel riquadro qui sotto ho scritto un semplicissimo programma in Phyton dove potete vedere cosa succede alla parabola cambiando il valore e/o il segno di $a,b$ e $c$_. Potete cambiare i valori di 3 diverse parabole per capire meglio cosa succede a ogni singola parabola per ogni singolo parametro.

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(-5,5,0.005) #!non modificare!#

a=1
b=1
c=1

y=a*x**2+b*x+c #!non modificare!#

plt.figure()
plt.grid()
plt.plot(x,y)
plt.show()
```

Qui invece ho creato un programma dove potete modificare tre parabole (rossa, verde e nera) contemporaneamente

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(-5,5,0.005) #!non modificare!#

##     prima parabola  in rosso     ###

a1=1
b1=1
c1=1
y1=a1*x**2+b1*x+c1 #!non modificare!#

##      seconda parabola in verde       ###

a2=-1
b2=1
c2=1
y2=a2*x**2+b2*x+c2 #!non modificare!#

##      terza parabola in nero       ###

a3=1
b3=1
c3=-1
y3=a3*x**2+b3*x+c3 #!non modificare!#


plt.figure()
plt.grid()
plt.plot(x,y1,color='red')
plt.plot(x,y2,color='green')
plt.plot(x,y3,color='black')
plt.show()
```
