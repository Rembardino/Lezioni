# Parabola 

La tipica formula di una parabola è  

<b><font size="4">$y=ax^2+bx+c$</font></b>

dove $a,b$ e $c$ sono numeri positivi o negativi. Il valore e il segno di questi tre parametri definiscono la posizione della parabola nello spazio. 

<font size="4">$a$</font>: il segno di $a$ cambia il verso della concavitá:
* se positivo ($a$ > 0) la concavitá è rivolta verso l'alto

* se negativo ($a$ < 0) la concavitá è rivolta verso il basso
* il valore assoluto di $a$ (modulo) indica quanto la parabola sia "schiacciata".  

<font size="4">$b$</font>: il coefficientte $b$ è legato alla posizione del vertice, del fuoco e dell'asse di simmetria.

<font size="4">$c$</font>: il coefficiente c indica il punto in cui la parabola interseca l'asse delle $y$


_Nel riquadro qui sotto ho scritto un semplicissimo programma in Phyton dove potete vedere cosa succede alla parabola cambiando il valore e/o il segno di $a,b$ e $c$_. Potete cambiare i valori di 3 diverse parabole per capire meglio cosa succede a ogni singola parabola per ogni singolo parametro.

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(-5,5,0.005) #!non modificare!#

a=1
b=0
c=0

y=a*x**2+b*x+c #!non modificare!#

plt.figure()
plt.grid()
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

<font size="4"> Ora che abbiamo capito come funziona la parabola, ne possiamo dare una definizione più formale.<font>

"_La parabola è un insieme di punti che hanno la stessa distanza da una retta <b>d</b> detta <b>direttrice,</b> e da un punto <b>F</b> detto <b>Fuoco</b>._
<br>
<br><br>
    
Per ogni Parabola possiamo definire le seguenti caratteristiche geometriche</b>
* <b>_l'asse di simmetria</b>:_ è la retta che divide in due parti uguali la parabola
* <b>_il vertice</b>:_ è il punto di intersezione tra l'asse di simmetria e la parabola.

<font size="4">Molti degli esercizi chiederanno di trovare le coordinate dei vari punti speciali della parabola. Queste dipendono dai valori dei parametri $a,b$ e $c$ dell'equazione della retta. Per risolvere gli esercizi dobbiamo introdurre una quantitá importante che tornerá utile anche in seguito per le equazioni di secondo grado: il discriminante ($\Delta$) 
  
$\Large\Delta=b^2-4ac$
<br><br>
    
    
Ora possiamo vedere quali sono le formule da usare per calcolare le coordinate $x$ e $y$ del fuoco e del vertice e le equazione dell'asse di simmetria e della retta direttrice.  <br><br><br>

<b>Fuoco:</b> $\Large\bigg(\frac{-b}{2a},\frac{1-\Delta}{4a}\bigg)$
    
<b>Vertice:</b> $\Large\bigg(\frac{-b}{2a},-\frac{\Delta}{4a}\bigg)$
    
    
<b>Equaione asse simmetria:</b> $\Large x=-\frac{b}{2a}$
    
    
<b>Equaione retta direttrice:</b> $\Large y=-\frac{1+\Delta}{4a}$
    
    
    

    
    







<font size="8"> Esempio: </font>
    
<font size="4">Data l'equazione della retta 

<font size="5">$y=x^2-4x+3$ </font>
<br> <br>
trovare il fuoco, il vertice, l'equazione dell'asse di simmetria e l'equazione della retta direttrice della parabola: <br> <br>
i valori dei parametri sono $a=1$, $b=-4$ e $c=3$
<br> <br>

    
$\large\Delta=b^2-4ac$ -----> $\large\Delta=(-4)^2-4*(1)*(3)$
    <br> <br>
    

<b>Fuoco:</b> $\Large\bigg(\frac{-b}{2a},\frac{1-\Delta}{4a}\bigg)$  -----> $\Large\bigg(\frac{-(-4)}{2*(1)},\frac{1-[(-4)^2-4*(1)*(3)]}{4*(1)}\bigg)$ -----> $\Large\bigg(2,-\frac{3}{4}\bigg)$
    <br> <br>
    
<b>Vertice:</b> $\Large\bigg(\frac{-b}{2a},-\frac{\Delta}{4a}\bigg)$  -----> $\Large\bigg(\frac{-(-4)}{2*(1)},-\frac{[(-4)^2-4*(1)*(3)]}{4*(1)}\bigg)$ -----> $\Large\bigg(2,-1\bigg)$
    
    
<b> Equaione asse simmetria:</b> $\Large x=-\frac{b}{2a}$  -----> $\Large x=-\frac{-4}{2*(1)}$ -----> $\Large x=2$
    
    
<b>Equaione retta direttrice:</b> $\Large y=-\frac{1+\Delta}{4a}$  -----> $\Large y=-\frac{1+[(-4)^2-4*(1)*(3)]}{4*(1)}$ -----> $\Large y=-\frac{5}{4}$  



```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)

x=np.arange(-1,5,0.005) #!non modificare!#

a=1
b=-4
c=3

y=a*x**2+b*x+c #!non modificare!#

plt.figure()
plt.grid(linewidth=2)
plt.plot(x,y)
plt.scatter(2,-3/4,color='red')
plt.scatter(2,-1,color='k')
plt.axvline(x=2,linestyle='dashed',color='purple',alpha=0.5)
plt.axhline(y=-5/4,color='green',)
plt.xlim(0,4)
plt.ylim(-2,2)
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```
