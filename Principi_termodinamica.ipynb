# Principi della Termodinamica

<font size="4"> Cerchiamo di capire subito cosa è la termodinamica. Notiamo innanzitutto che c'è la parola dinamica nella parola, quindi ci aspettiamo un movimento di qualcosa. L'altra parola è "Termo" che è subito associabile alla temperatura. Quindi possiamo dire che la Termodinamica studia il movimento della temperatura. E in effetti è proprio cosi. Ora resta da capire cosa sia la temperatura.

<font size="4"> Ogni giorno tocchiamo oggetti con diverse temperature e ci accorgiamo subito della differenza. Ma cosa è che fa la differenza tra un corpo freddo e uno caldo? Quello che cambia è la velocitá con cui si muovono gli atomi all'interno del corpo. Quindi possiamo pensare che un termometro o un termo-scacnner non sono nient'altro che autovelox per la velocità degli atomi. Pensando alla temperatura in questo modo, ci risulterv molto più facile capire i prossimi concetti. <br><br>
Per esempio possiamo capire del perchè nel sistema internazionale di misura la Temperatura si misura in Kelvin [K]  e che la temperatura minima è 0 gradi Kelvin (lo 0 assoluto). Questo perchè la temperatura è la misura della velocità degli atomi e la velocità minima è 0! <br><br>
Come abbiamo fatto per la dinamica, possiamo sempre definire l'energia cinetica del corpo associata alla velocità come $E_k=\frac{1}{2}*m*v^2$ dove v è la velocità.    <br><br> 
Dopo questa breve introduzione, vediamo i principi della termodinamica. Come sempre, li enunciamo tutti quanti e poi li vediamo nel dettaglio uno per uno.

<font size="6"> <b> Principio zero </font><br>
    <font size="4"> _Il principio 0 ci dice che se il corpo A è in equilibrio termico con il corpo B, e il corpo B è in equilibrio termico con il corpo C, allora il corpo A è in equilibrio termico con il corpo C._

<font size="4">Che cosa vuol dire equilibrio termico? Semplicemente che due corpi hanno la stessa temperatura. Una volta capito questo risulta immediato capire che se il corpo A ha la stessa temperatura del corpo B, e il corpo B ha la stessa temperatura del corpo C, allora il corpo A ha la stessa temperatura del corpo C. <br> <br>
_Ma come si arriva all'equilibrio? Che succede se metto in contatto due corpi con diversa temperatura?_ <br><br>
Cosa succcede quando durante l'inverno tocchiamo il termosifone per scaldarci? Succede che cerchiamo di andare in equilibrio termico con il termosifone. Noi siamo freddi e "rubiamo" il caldo dal termosifone.<br> <br>
In estate invece per rifrescarci magari ci mettiamo sulla fronte una bella bottiglia di acqua gelata per "cedere" un pochino della nostra temperatura alla bottiglia che si scalda.<br> <br>
Questo trasferimento di temperatura lo viviamo ogni giorno quindi non ci ragioniamo troppo su cosa stia veramente accadendo. Perchè quando tocchiamo il termosifone siamo noi che rubiamo il calore da lui e non il contrario? E perchè quando ci mettiamo la bottiglietta fredda è lei che ci ruba il calore e non noi che "rubiamo" il freddo? <br> <br>
Quello che succcede è che quando due corpi entrano in contatto, tendono ad arrivare alla stessa temperatura (equilibrio termico) e per farlo <font color='red'> _il corpo più caldo passa (spontaneamente) il calore al corpo più freddo_ e <b>MAI</b> il contrario.</font>

Quindi il termosifone cede la temperatura a noi come noi la cediamo alla bottiglia di acqua gelata fino a quando non raggiungiamo l'equilibrio con il corpo. Ma questo vuol dire che la velocitá degli atomi del corpo più caldo diminuisce mentre quella del corpo più freddo aumenta.



   

<font size="6
            "> <b> Primo principio </font><br> <font size="4"> _L'energia interna di un sistema termodinamico isolato è costante_

<font size="4"> Questo principio ci ricorda molto il principio della conservazione dell'energia meccanica. In effetti è mooolto simile. Nella conservzione dell'energia meccanica avevamo che se essa si conservava, voleva dire che l'energia si trasformava in qualche modo. Per esempio l'energia cinetica si trasformava in energia potenziale o viceversa. Se nel sistema erano presenti molle, l'energia si poteva anche trasformare in energia elastica. Nella termodinamica vale lo stesso principio. Ricordiamoci sempre del principio di Lavoisier:<br><br>
 _In natura, nulla si crea, nulla si distrugge ma tutto si trasforma_ <br><br>

<font size="4"> Nel mondo termodinamico questo principio si traduce nella seguente formula: <br><br>
    <b><font size="6">
    $\Delta U=Q-L$</font></b> <br><br>
Dove U è l'energia interna (contenuta) del sistema (somma di tutte le energie), Q è il calore e L è il lavoro. L'unica grandezza nuova in questa formula è (dovrebbe essere) il calore. Infatti l'energia interna è la somma di tutte le energie (interne) del sistema e il lavoro si riferisce proprio al Lavoro delle forze (ricordo che il Lavoro è il prodotto tra la forza e lo spostamento).
Questa formula ci sta dicendo che se noi osserviamo un qualche cambiamento($\Delta$) dell'energia interna del sistema, allora dobbiamo aver compiuto del Lavoro o scambiato calore. <br> <br>
Se ci concentriamo solamente sul Lavoro per adesso, ci accorgiamo che questa formula segue lo stesso concetto del teorema dell'energia cinetica che ci diceva che la differenza di energia cinetica è proprio il lavoro compiuto!! 
<br> <br> Concentriamoci ora sulla nuova grandezza Q. Che cosa è il calore? Il calore è tutta l'energia persa del sistema che non si è trasformata in lavoro. Non ci avete capito niente vero?, ve lo ridico allora.
    
Torniamo all'esempio della bottiglietta di acqua gelata e la nostra fronte in estate. Noi abbiamo caldo perchè la nostra energia interna è molto alta. Per questo motivo ci mettiamo la bottiglietta sulla fronte e ci sentiamo bene perchè la nostra energia interna è diminuita. Ma perchè è diminuita? Perche abbiamo _scambiato_ calore con la bottiglietta. La nostra energia interna è diminuita di una quantitá pari al calore che abbiamo trasferito alla bottiglietta. Durante questo passaggio, non c'è stato nessuno spostamento <b> quindi </b> nessun lavoro ma la nostra energia interna si è _trasformata_ in calore _scambiato_. <br> <br>
Molto probabilmente, negli esercizi di dinamica con l'attrito avete sempre detto che l'energia si è dispersa anche in calore ma (forse) senza ben capire da dove venisse questa parola e questo concetto. Bene ora lo sapete. Se sfregate il vostro dito sul tavolo, l'attrito compie lavoro (c'è lo spostamento del vostro dito) ma potete ben sentire che c'è amche uno scambio di calore visto che il vostro dito si scalda! bene questi sono L e Q.<br> <br>
    
Voglio sottolineare che fino a questo momento non è stata fatta nessuna semplificazione! Tutto questo è vero sempre! Ovviamente con sistemi complicati è difficile definire l'energia interna, il lavoro e il calore.

<font size="6"> <b> Secondo principio </font><br> <font size="4"> _In un sistema isolato, l'entropia totale del sistema non può diminuire_

<font size="4"> È finalmente giunto il momento di nominare l'Entropia. Sicuramente avete sentito la seguente frase:"l'entropia misura il grado di disordine del sistema". Questo è sicuramente vero, ma lasciatemi dire che l'entropia non è solo questo. Vedremo in seguito la vera definizione di entropia. <br> <br>
    Per adesso vediamo la formula del secondo principio: <br> <br>
 <b><font size="6">
    $\Delta S\geqslant0 $</font></b> <br><br>
    Questo vuol dire che durante qualsiasi trasformazione che avviene in un sistema isolato l'entropia <b>totale </b> non può <b>diminuire</b>. Notiamo che il simbolo è maggiore uguale, quindi la variazione di entropia può essere 0, ma non può essere negativa. Possiamo riscrivere la formula come:  <br> <br> <b><font size="6"> $S_{finale}-S_{iniziale}\geqslant0 $</font></b><br><br>
Se consideriamo l'intero universo come un sistema isolato (cosa che possiamo assumere vera) allora ci accorgiamo che il secondo principio della termodinamica è LA LEGGE che move il sole e l'altre stelle (cit. Dante). Tutto quello che accade nell'universo accade seguendo il secondo principio della termodinamica. Ovvero aumentando l'entropia fino al suo massimo valore possibile! <br><br>
    
<font size="5"> _Ma che succede allora quando si arriva all'entropia massima?_   
    

<details><summary>Cliccare qui per la soluzione</summary>

 <font size="4"> Ci troviamo in uno stato di equilibrio. Quando ci mettiamo la bottiglietta fredda in fronte per rifrescarci, scambiamo calore con la bottiglia, passiamo la nostra energia alla bottiglia. Durante questo processo l'entropia totale aumenta. Nel momento in cui la nostra fronte e la bottiglietta hanno la stessa temperatura, sono in equilibrio termico, si è raggiunta la massima entropia. <br><br>
     
 
</details>

 <font size="4"> Il fatto che il calore passa dal corpo più caldo al corpo più freddo è dettato dal secondo principio, se cosi non fosse, si avrebbe una diminuzione dell'entropia totale che è impossibile. <br><br>
    Ora dovrebbe sorgere una domanda... _Ma se vale il secondo principio, quindi il calore passa da un corpo caldo a un corpo freddo, come è possibile che i i cibi si congelano nel congelatore e sono freschi nel frigo?_ <br><br>
 
La risposta è semplice: Perchè stiamo fornendo energia sul sitema tramite il motore del congelatore. Se sommiamo tutte le entropie dei sistemi, ci accorgiamo che l'entropia aumenta. Quindi il calore può passare da un corpo caldo a uno più freddo, ma non SPONTANEAMENTE! Per far si che accada, bisogna spendere energia al sistema. S

## Gas ideale o perfetto:

<font size="4"> Di solito la termodinamica è associata allo studio dei gas, ma spero di avervi convinto che la termodinamica coinvolge tutto l'universo. Per questione di praticitá, negli esercizi si introduce spesso il gas perfetto che è un gas con le seguenti caratteristiche:<br><br>
* le molecole sono puntiformi e pertanto hanno un volume trascurabile<br><br>
* gli urti tra le particelle el il contenitore sono perfettamente elastici (si conserva energia e quantida di moto)<br><br>
* le particelle non interagiscono tra di loro<br><br>
* le particelle sono indistinguibili tra di loro.<br><br>

Se tutto questo è vero, allora il gas soddisfa la legge dei gas perfetti:<br><br>
    <b><font size="6">
    $pV=NRT$</font></b> <br><br>
dove p è la pressione del gas [si misura in Pascal, Pa], V è il volume dove è contenuto il gas, N è il numero delle particelle che costituiscono il gas, R è la costante dei gas [=8,314462618... (J/mol)*K] e T è la temperatura [si misura in gradi Kelvin, K].

## Trasformazioni reversibili:


<font size="4"> Il primo concetto fondamentale è quello di trasformazione reversibile. Se passo dallo stato A allo stato B con una serie infinita di piccoli steps, allora la trasformazione si dice reversibile. Questo vuol dire che posso tornare indietro senza problemi dallo stato B allo stato A. Se questo non è possibile, allora la trasformazione è irreversibile.
